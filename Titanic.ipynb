```python
train['Age']=train['Age'].replace(np.NaN,train['Age'].mean())
```

```python
train['Embarked']=train['Embarked'].replace(np.NaN,'C')
```

```python
for i in range(891):
    if train['Age'].values[i]<=5:
        train['Age'].values[i]=0
    elif train['Age'].values[i]<=10 and train['Age'].values[i]>5 :
        train['Age'].values[i]=1
    elif train['Age'].values[i]<=20 and train['Age'].values[i]>10:
        train['Age'].values[i]=2
    elif train['Age'].values[i]<=30 and train['Age'].values[i]>20:
        train['Age'].values[i]=3
    elif train['Age'].values[i]<=40 and train['Age'].values[i]>30:
        train['Age'].values[i]=4
    elif train['Age'].values[i]<=50 and train['Age'].values[i]>40:
        train['Age'].values[i]=5
    elif train['Age'].values[i]<=60 and train['Age'].values[i]>50:
        train['Age'].values[i]=6
    elif train['Age'].values[i]<=70 and train['Age'].values[i]>60:
        train['Age'].values[i]=7
    else:
        train['Age'].values[i]=8
```

```python
train['Fare'].max()
```

```python
for i in range(891):
    if train['Fare'].values[i]<=5:
        train['Fare'].values[i]=0
    elif train['Fare'].values[i]>5 and train['Fare'].values[i]<=10:
        train['Fare'].values[i]=1
    elif train['Fare'].values[i]>10 and train['Fare'].values[i]<=15:
        train['Fare'].values[i]=2
    elif train['Fare'].values[i]>15 and train['Fare'].values[i]<=20:
        train['Fare'].values[i]=3
    elif train['Fare'].values[i]>20 and train['Fare'].values[i]<=25:
        train['Fare'].values[i]=4
    elif train['Fare'].values[i]>25 and train['Fare'].values[i]<=30:
        train['Fare'].values[i]=5
    elif train['Fare'].values[i]>30 and train['Fare'].values[i]<=35:
        train['Fare'].values[i]=6
    elif train['Fare'].values[i]>35 and train['Fare'].values[i]<=40:
        train['Fare'].values[i]=7
    elif train['Fare'].values[i]>40 and train['Fare'].values[i]<=45:
        train['Fare'].values[i]=8
    elif train['Fare'].values[i]>45 and train['Fare'].values[i]<=50:
        train['Fare'].values[i]=9
    elif train['Fare'].values[i]>50 and train['Fare'].values[i]<=75:
        train['Fare'].values[i]=10
    elif train['Fare'].values[i]>75 and train['Fare'].values[i]<=100:
        train['Fare'].values[i]=11
    elif train['Fare'].values[i]>100 and train['Fare'].values[i]<=200:
        train['Fare'].values[i]=12
    elif train['Fare'].values[i]>200 and train['Fare'].values[i]<=300:
        train['Fare'].values[i]=13
    else: 
         train['Fare'].values[i]=14
    
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(6,input_dim=6,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(5,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
y_train
```

```python
history=model.fit(train,y_train,validation_split=0.1,epochs=10)
```
